```python
loss_acc = open("./loss_graph.txt", 'r')
loss_acc = loss_acc.read().splitlines()

loss_acc
```

```python
loss_acc_clean = []

for line in loss_acc:
    if 'Batch' not in line and line != '' and 'Time' not in line and 'Saving' not in line:
        loss_acc_clean.append(line)

loss_acc_clean
```

```python
losses = []
accuracies = []

for line in loss_acc_clean:
    l = line[line.find('ss') + 3:]
    l = l[:6]
    
    a = line[line.find('cy') + 3:]
    
    losses.append( float(l) )
    accuracies.append( float(a) )
```

```python
import matplotlib.pyplot as plt
import numpy as np

epochs = [i for i in range(1, 101)]
plt.plot(epochs, losses, color="orange", linewidth=3)
plt.ylim(0, 6)
plt.yticks(np.arange(min(losses)-.2, max(losses)+1, 1.0))

plt.title('Sparse Categorical Crossentropy Loss of Transformer Model')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

```python
plt.plot(epochs, accuracies, color="orange", linewidth=3)

plt.title('Sparse Categorial Accuracy of Transformer Model')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.show()
```
